```python
print(benin.isnull().sum())
print(sirraleon.isnull().sum())
print(togo.isnull().sum())

# Checking for outliers and anomalies
import numpy as np

def detect_outliers(df, column):

    # Z-score method
    
    z_scores = np.abs(stats.zscore(df[column].dropna()))
    return df[(z_scores > 3)]


outliers_benin = detect_outliers(benin, 'GHI')
outlier_sierraleone = detect_outliers(sirraleon, 'GHI')
outlier_togo = detect_outliers(togo, 'GHI')
print(outliers_benin)
print(outlier_sierraleone)
print(outlier_togo)
```
